```python
import pandas as pd
```

```python
raw_data = pd.read_csv("/content/GDSC2_public_raw_data_27Oct23.csv")
```

```python
raw_cell_line_names = raw_data["CELL_LINE_NAME"].unique()
len(raw_cell_line_names)
```

```python
cosmic_ids_raw = list(set(raw_data.COSMIC_ID.tolist()))
```

```python
from glob import glob
cg_file_paths = glob('/content/CellLines_CG_BEMs/*_SEQ_BEM.txt')  # Adjust path as necessary

unique_ids = set()

# Process each file
for file_path in cg_file_paths:
    cg_data = pd.read_csv(file_path, sep='\t')
    cell_line_ids = cg_data.columns[2:]  # Skip first two columns (assuming those are not IDs)
    unique_ids.update(cell_line_ids)  # Add to set to ensure uniqueness

# Count of unique cell line IDs across all files
unique_cell_line_count = len(unique_ids)

cg_data_frames = []

for file_path in cg_file_paths:
    data = pd.read_csv(file_path, sep='\t').set_index('CG')
    cg_data_frames.append(data)

combined_cg_data = pd.concat(cg_data_frames, axis=1)

combined_cg_data = combined_cg_data.groupby(combined_cg_data.columns, axis=1).max()

combined_cg_data = combined_cg_data.T
combined_cg_data_cleaned = combined_cg_data[combined_cg_data.index.str.isnumeric()]

```

```python
cnv_file_paths = glob('/content/CellLine_CNV_BEMs/*_CNA_BEM.rdata.txt')  # Adjust path as necessary

unique_ids = set()

cnv_data_frames = []

for file_path in cnv_file_paths:
    data = pd.read_csv(file_path, sep='\t').set_index('Unnamed: 0')
    cnv_data_frames.append(data)
```

```python
combined_cnv_data = pd.concat(cnv_data_frames, axis=1).fillna(0)
combined_cnv_data
```

```python
combined_cnv_data = combined_cnv_data.groupby(combined_cnv_data.columns, axis=1).max()
combined_cnv_data
```

```python
meth_file_paths = glob('/content/METH_CELLLINES_BEMs/*.txt')  # Adjust path as necessary

unique_ids = set()

# Process each file
for file_path in meth_file_paths:
    meth_data = pd.read_csv(file_path, sep='\t')
    cell_line_ids = meth_data.columns[2:]  # Skip first two columns (assuming those are not IDs)
    unique_ids.update(cell_line_ids)  # Add to set to ensure uniqueness

# Count of unique cell line IDs across all files
unique_cell_line_count = len(unique_ids)

meth_data_frames = []

for file_path in meth_file_paths:
    data = pd.read_csv(file_path, sep='\t').set_index('Unnamed: 0')
    meth_data_frames.append(data)

combined_meth_data = pd.concat(meth_data_frames, axis=1).fillna(0)
```

```python
combined_meth_data = combined_meth_data.groupby(combined_meth_data.columns, axis=1).max()

```

```python

combined_meth_data = combined_meth_data.T
combined_meth_data_cleaned = combined_meth_data[combined_meth_data.index.str.isnumeric()]
```

```python
combined_meth_data_cleaned
```

```python
combined_cg_data_cleaned.to_csv("/content/binary_cg_data.csv")
combined_cnv_data.to_csv("/content/binary_cnv_data.csv")
combined_meth_data_cleaned.to_csv("/content/binary_meth_data.csv")
```

```python
print("cg_data: ", combined_cg_data_cleaned.shape)
print("cnv_data: ", combined_cnv_data.shape)
print("meth_data: ", combined_meth_data_cleaned.shape)
```

```python
id_to_keep = list(set(combined_cg_data_cleaned.index))
id_to_keep = set(map(int, id_to_keep))
```

```python
combined_cnv_data.index = combined_cnv_data.index.astype(int)
cnv_data_filtered = combined_cnv_data.loc[combined_cnv_data.index.intersection(id_to_keep)]
cnv_data_filtered

id_kept = list(set(cnv_data_filtered.index))
id_kept = set(map(int,id_kept))
combined_cg_data_cleaned.index = combined_cg_data_cleaned.index.astype(int)
cg_data_filtered = combined_cg_data_cleaned.loc[combined_cg_data_cleaned.index.intersection(id_kept)]
```

```python
combined_meth_data_cleaned.index = combined_meth_data_cleaned.index.astype(int)
meth_data_filtered = combined_meth_data_cleaned.loc[combined_meth_data_cleaned.index.intersection(id_kept)]

meth_data_filtered
```

```python
print("cg_data: ", cg_data_filtered.shape)
print("cnv_data: ", cnv_data_filtered.shape)
print("meth_data: ", meth_data_filtered.shape)
```

```python
combined_data_DNN = pd.concat([cg_data_filtered, cnv_data_filtered], axis=1)
```

```python
def transform_binary_matrix(df):
    return df.applymap(lambda x: 1 if x == 1 else (-1 if x == 0 else 0))

combined_data_DNN = transform_binary_matrix(combined_data_DNN)
```

```python
meth_data_filtered = transform_binary_matrix(meth_data_filtered)
```

```python
combined_data_DNN
```

```python
meth_data_aligned = meth_data_filtered.reindex(index=combined_data_DNN.index, fill_value=0)
combined_data_DNN = pd.concat([combined_data_DNN, meth_data_aligned], axis=1)

```

```python
combined_data_DNN
```

```python
#matching_indices_count = len(set(combined_data_DNN.index).intersection(cosmic_ids_raw))

matched_data = combined_data_DNN.loc[combined_data_DNN.index.intersection(cosmic_ids_raw)]

# Verify the shape to confirm 925 matching rows
matched_data.shape
```
